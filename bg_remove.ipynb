```python
import cv2
import os
import glob
from matplotlib import pyplot as plt
import numpy as np
import skimage.color
```

```python
images = [cv2.imread(file) for file in glob.glob("C:/Users/Prajwol/RiceDisease/images/*.jpg")]
```

```python
l = len(images)
type(images)
```

```python
for x in range(0, l):
    hsv_img = cv2.cvtColor(images[x], cv2.COLOR_BGR2HSV)
    s = hsv_img[:, :, 1]
    sigma = 1
    t = 0.45
    blur = skimage.color.rgb2gray(s)
    blur = skimage.filters.gaussian(blur, sigma=sigma)
    mask = blur > t
    rgb_img = cv2.cvtColor(hsv_img, cv2.COLOR_HSV2RGB)
    sel = np.zeros_like(rgb_img)
    sel[mask] = rgb_img[mask]
    skimage.io.imsave("C:\\Users\\Prajwol\\RiceDisease\\bgremove\\bgremove_"+str(x)+".jpg", sel)
```
